```python
import csv

dados_database = {'ID da Venda': [], 'Cancelado': [], 'Nome do Produto': [], 'Valor do Produto': [], 'Quantidade Vendida': [], 'Nome do Vendedor': [], 'CPF do Comprador': [], 'Data da Venda': [], 'Informações Adicionais': []} # Inicializa um dicionário vazio para a base de dados

def bem_vindo():
    '''
    Inicializa o gerenciador de vendas e direciona o usuário ao menu principal.
    '''

    print('#####################################')
    print('##### GERENCIADOR DE VENDAS ADA #####')
    print('#####################################')

    principal()

def navegador_menus(menu: dict):
    '''
    Função utilizada para invocar outras funções dentro do código, baseado no input do usuário.
    '''

    print('Opções disponíveis:')
    for chave in menu:
        print(f'[{chave}] - {menu[chave][0]}')
    print('\n')
    opcao_selecionada = valida_opcao(menu)      # Invoca a função valida_opcao para verificar o input de dado do usuário
    menu[opcao_selecionada][1]()                # Com base na opção selecionada, invoca outras funções conforme dicionário do menu atual

def valida_opcao(menu: dict) -> int:
    '''
    Função utilizada para verificar se o input do usuário atende aos valores do menu de navegação atual.
    '''

    while True:
        opcao_selecionada = input('\nDigite o número da opção desejada: ')

        try:
            opcao_selecionada = int(opcao_selecionada)
            if opcao_selecionada > len(menu) or opcao_selecionada <= 0:
                print('\nOpção inválida! Selecione apenas entre os valores disponíveis!')
                continue
            else:
                return opcao_selecionada
        except:
            print('\nEntrada inválida! Utilize apenas números inteiros!')
            continue

def principal():
    '''
    Menu principal de navegação no programa.
    '''

    print('\n##### MENU PRINCIPAL #####\n')

    navegador_menus(menu_principal)

def gerencia_vendas():
    '''
    Menu para inclusão e exclusão de vendas realizadas.
    '''

    print('##### GERENCIADOR DE VENDAS #####\n')
    
    navegador_menus(menu_vendas)

def informacoes_gerais():
    '''
    Menu para informações sobre faturamento, equipe e produtos.
    '''

    print('##### INFORMAÇÕES GERAIS #####\n')

    navegador_menus(menu_informacoes)

def ferramentas_adicionais():
    '''
    Menu para ferramentas de exportação de dados e conversão de moedas.
    '''

    print('##### FERRAMENTAS ADICIONAIS #####\n')

    navegador_menus(menu_ferramentas)


def cadastrar_venda():
    '''
    Inicializa input para inclusão de vendas realizadas.
    '''

    nomes_produtos = []
    valores_produtos = []
    quantidades_produtos = []
    while True:                         # Laço criado para inclusão de múltiplos produtos, valores e quantidades em uma única venda
        nome_produto = input('Digite o nome do produto adquirido: ')
        nomes_produtos.append(nome_produto)
        valor_produto = float(input('Digite o valor do produto, em Reais (use . para separar os decimais): '))
        valores_produtos.append(valor_produto)
        quantidade_produto = int(input('Digite a quantidade de produtos adquiridos: '))
        quantidades_produtos.append(quantidade_produto)

        continuar = input('Deseja inserir mais produtos nesta venda? [s/n]').lower()
        if continuar != 's':
            break
    
    vendedor = input('Digite o nome do vendedor: ')
    comprador = int(input('Digite o CPF do comprador (utilize apenas números): '))
    data = input('Digite a data da compra, usando o formato DD/MM/AAAA: ')
    adicionais = input('Digite as informações adicionais da venda: ')

    inclui_venda(dados_database, nomes_produtos, valores_produtos, quantidades_produtos, vendedor, comprador, data, adicionais)

def inclui_venda(database: dict, nomes_produtos: list, valores_produtos: list, quantidades_produtos: list, vendedor: list, comprador: list, data: list, adicionais: list):
    '''
    Adiciona a venda inserida pela função cadastrar_venda na base de dados.
    '''

    if len(database) == 0:
        database['ID da Venda'] = [1]
        database['Cancelado'] = [False]
        database['Nome do Produto'] = [nomes_produtos]
        database['Valor do Produto'] = [valores_produtos]
        database['Quantidade Vendida'] = [quantidades_produtos]
        database['Nome do Vendedor'] = [vendedor]
        database['CPF do Comprador'] = [comprador]
        database['Data da Venda'] = [data]
        database['Informações Adicionais'] = [adicionais]
    else:
        database['ID da Venda'].append(len(database['ID da Venda'])+1)
        database['Cancelado'].append(False)
        database['Nome do Produto'].append(nomes_produtos)
        database['Valor do Produto'].append(valores_produtos)
        database['Quantidade Vendida'].append(quantidades_produtos)
        database['Nome do Vendedor'].append(vendedor) 
        database['CPF do Comprador'].append(comprador) 
        database['Data da Venda'].append(data) 
        database['Informações Adicionais'].append(adicionais)
    
    print('Venda incluída com sucesso!\n')

    gerencia_vendas()

def consulta_venda():
    '''
    Consulta os dados de uma venda realizada através do input do ID da Venda.
    '''

    id_consulta = int(input('Digite o ID da Venda para consulta: '))
    if id_consulta in dados_database['ID da Venda']:
        idx = dados_database['ID da Venda'].index(id_consulta)
        for dado in dados_database:
            print(f'{dado}: {dados_database[dado][idx]}')
        print('\n')
    else:
        print(f'A venda de ID {id_consulta} nâo consta na base de dados!\n')
    
    gerencia_vendas()

def cancela_venda():
    '''
    Consta uma venda como cancelada através do input do ID da Venda.
    '''

    # NOTA: Optou-se apenas pela alteração do campo "Cancelado" para True ao invés de remover a linha da base de dados.
    #       Desta forma, a base completa valerá como histórico de todas as entradas do sistema, não sendo estes valores,
    #       entretanto, utilizados nos cálculos das médias e demais indicadores de vendas.

    id_cancelamento = int(input('Digite o ID da Venda para cancelar: '))
    if id_cancelamento in dados_database['ID da Venda']:
        idx = dados_database['ID da Venda'].index(id_cancelamento)
        confirmacao = input(f'Deseja mesmo cancelar a venda de ID {id_cancelamento}? [s/n]: ').lower()
        if confirmacao == 's':
            dados_database['Cancelado'][idx] = True
            print('Venda cancelada com sucesso!\n')
        else:
            print('Input inválido! Tente novamente!')
    else:
        print(f'A venda de ID {id_cancelamento} nâo consta na base de dados!\n')
    
    gerencia_vendas()

def consulta_ano_mes(ano_consulta: float, mes_consulta: float) -> list:
    '''
    Função utilizada pra filtrar o ano e mês selecionados pelo usuário, removendo as vendas canceladas
    '''
    
    datas_separadas = [data.split('/') for data in dados_database['Data da Venda']]     # Separador de mes e ano
    indices_consulta = []

    for idx, data in enumerate(datas_separadas):
        if int(data[2]) == int(ano_consulta):
            if int(data[1]) == int(mes_consulta):
                if dados_database['Cancelado'][idx]:        # Caso a venda tenha sido cancelada (True), o código passa pra próxima venda
                    continue
                else:
                    indices_consulta.append(idx)
    
    return indices_consulta

def faturamento_mensal():
    '''
    Função para informação do faturamento mensal.
    '''

    ano_consulta = input('Digite o ano para consuta: ')
    mes_consulta = input('Digite o número do mês para consulta: ')
    indices_consulta = consulta_ano_mes(ano_consulta, mes_consulta)
    soma = 0
    
    try:
        # A recursão abaixo utiliza o índice (idx) e sub índice (sub_idx) devido à possibilidade de múltiplas produtos vendidos em um único ID

        for idx in indices_consulta:
            for sub_idx in range(len(dados_database['Nome do Produto'][idx])):
                valor_parcial = dados_database['Valor do Produto'][idx][sub_idx] * dados_database['Quantidade Vendida'][idx][sub_idx]
                soma += valor_parcial
            
        print(f'O Valor Total de vendas em {mes_consulta}/{ano_consulta} foi de R${soma:.2f}, com Valor Médio por Venda de R${soma/len(indices_consulta):.2f}\n')
    except ZeroDivisionError:
        print(f'{mes_consulta}/{ano_consulta} não possui vendas!\n')

    informacoes_gerais()

def informacao_equipe():
    '''
    Função para obtenção do vendedor com maior valor total de vendas realizadas no mês e ano indicado.
    '''

    ano_consulta = input('Digite o ano para consuta: ')
    mes_consulta = input('Digite o número do mês para consulta: ')
    indices_consulta = consulta_ano_mes(ano_consulta, mes_consulta)

    vendedores_individuais = [individuo for idx, individuo in enumerate(dados_database['Nome do Vendedor']) if individuo not in dados_database['Nome do Vendedor'][:idx]]
    total_por_vendedor = []

    # A recursão abaixo utiliza o índice (idx) e sub índice (sub_idx) devido à possibilidade de múltiplas produtos vendidos em um único ID

    for vendedor in vendedores_individuais:
        soma = 0
        for idx in indices_consulta:
            if dados_database['Nome do Vendedor'][idx] == vendedor:
                for sub_idx in range(len(dados_database['Nome do Produto'][idx])):
                    valor_parcial = dados_database['Valor do Produto'][idx][sub_idx] * dados_database['Quantidade Vendida'][idx][sub_idx]
                    soma += valor_parcial
        total_por_vendedor.append(soma)
    
    maior_valor = max(total_por_vendedor)
    maior_vendedor = vendedores_individuais[total_por_vendedor.index(maior_valor)]

    if maior_valor == 0:
        print(f'{mes_consulta}/{ano_consulta} não possui vendas cadastradas!\n')
    else:
        print(f'O maior vendedor de {mes_consulta}/{ano_consulta} foi {maior_vendedor}, totalizando R${maior_valor:.2f} em vendas!\n')

    informacoes_gerais()

def informacao_produto():
    '''
    Função para obter o produto com maior venda no mês e ano solicitado, em quantidade.
    '''

    ano_consulta = input('Digite o ano para consuta: ')
    mes_consulta = input('Digite o número do mês para consulta: ')
    indices_consulta = consulta_ano_mes(ano_consulta, mes_consulta)
    produtos_vendidos = {}

    # A recursão abaixo utiliza o índice (idx) e sub índice (sub_idx) devido à possibilidade de múltiplas produtos vendidos em um único ID

    try:
        for idx in indices_consulta:
            for sub_idx in range(len(dados_database['Nome do Produto'][idx])):
                produto = dados_database['Nome do Produto'][idx][sub_idx]
                quantidade = dados_database['Quantidade Vendida'][idx][sub_idx]
                if produto not in produtos_vendidos:
                    produtos_vendidos[produto] = quantidade
                else:
                    produtos_vendidos[produto] += quantidade

        maior_produto = max(produtos_vendidos)
        maior_quantidade = produtos_vendidos[maior_produto]

        print(f'O produto mais vendido em {mes_consulta}/{ano_consulta} foi {maior_produto}, que teve {maior_quantidade} unidades vendidas!\n')
    except ZeroDivisionError:
        print(f'{mes_consulta}/{ano_consulta} não possui vendas!\n')
    except ValueError:
        print(f'Nenhum dado encontrado para {mes_consulta}/{ano_consulta}!\n')

    informacoes_gerais()

def exporta_dados():
    '''
    Função para exportação de arquivo CSV da base de dados.
    '''

    # A função lambda abaixo é utilizada apenas na criação de dicionário contendo as chaves da base de dados para extração do CSV

    extrator_dicionario = lambda dicionario: [chave for chave in dicionario.keys()]
    colunas = extrator_dicionario(dados_database)

    try:
        with open('./vendas.csv', 'w') as arquivo:
            escritor = csv.DictWriter(arquivo, fieldnames=colunas, lineterminator='\n')
            escritor.writeheader()
            for idx in range(len(dados_database['ID da Venda'])):
                linha = {}
                for coluna in colunas:
                    linha[coluna] = dados_database[coluna][idx]
                escritor.writerow(linha)
                
    except Exception as erro:
        print(f'O programa retornou o erro {erro}')
    else:
        print('O arquivo foi exportado com sucesso!')
    finally:
        print('Verifique o arquivo gerado na pasta de execução deste script.\n')

    ferramentas_adicionais()

def converte_moedas():      # Utilizada uma Função Geradora, conforme solicitado no projeto
    '''
    Função para conversão dos valores dos produtos em dólar, conforme cotação atual.
    '''

    # A função lambda abaixo é utilizada apenas na criação de dicionário contendo as chaves da base de dados para extração do CSV

    extrator_dicionario = lambda dicionario: [chave for chave in dicionario.keys()]
    colunas = extrator_dicionario(dados_database)

    cambio = float(input('Digite o valor atual do Dólar, em reais, utilizando . como separador de decimais: '))
    dados_conversao = dict(dados_database)

    yield 'A função irá agora converter os valores da base de dados!'
    
    dolar = lambda x: x/cambio      # Função lambda utilizada para converter os valores de reais para dólares.
    
    for idx in range(len(dados_conversao['ID da Venda'])):
        valor_convertido = []
        for produto in dados_conversao['Valor do Produto'][idx]:
            novo_valor = dolar(produto)
            valor_convertido.append(novo_valor)
        dados_conversao['Valor do Produto'][idx] = valor_convertido

    # print(dados_conversao)

    yield 'A função irá agora exportar o arquivo CSV!'

    try:
        with open('./vendas_dolares.csv', 'w') as arquivo:
            escritor = csv.DictWriter(arquivo, fieldnames=colunas, lineterminator='\n')
            escritor.writeheader()
            for idx in range(len(dados_conversao['ID da Venda'])):
                linha = {}
                for coluna in colunas:
                    linha[coluna] = dados_conversao[coluna][idx]
                escritor.writerow(linha)
                
    except Exception as erro:
        print(f'O programa retornou o erro {erro}')
    else:
        print('O arquivo foi exportado com sucesso!')
    finally:
        print('Verifique o arquivo gerado na pasta de execução deste script.\n')
    
    ferramentas_adicionais()

def conversor():
    '''
    Função para invocar recursivamente a função geradora converte_moedas.
    '''
    
    conversao = converte_moedas()
    for passo in conversao:
        print(passo)

def sair_menu():
    '''
    Função para finalizar a execução do progama, presente nos menus de navegação.
    '''

    print('\nObrigado por utilizar o Gerenciador de Vendas ADA!')

# Abaixo lista-se os menus disponíveis para o usuário

menu_principal = {1: ('Gerenciador de Vendas', gerencia_vendas), 
                  2: ('Informações Gerais', informacoes_gerais), 
                  3: ('Ferramentas Adicionais', ferramentas_adicionais), 
                  4: ('Sair do Programa', sair_menu)}

menu_vendas = {1: ('Incluir Venda', cadastrar_venda), 
               2: ('Consultar Vendas', consulta_venda), 
               3: ('Cancelar Venda', cancela_venda), 
               4: ('Voltar ao Menu Principal', principal), 
               5: ('Sair do Programa', sair_menu)}

menu_informacoes = {1: ('Faturamento Mensal', faturamento_mensal),
                    2: ('Informações de Equipe', informacao_equipe),
                    3: ('Informações de Produto', informacao_produto),
                    4: ('Voltar ao Menu Principal', principal),
                    5: ('Sair do Programa', sair_menu)}

menu_ferramentas = {1: ('Exportar Dados', exporta_dados),
                    2: ('Conversor de Moedas', conversor),
                    3: ('Voltar ao Menu Principal', principal),
                    4: ('Sair do Programa', sair_menu)}
```
